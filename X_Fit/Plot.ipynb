```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = 'rms.plot'
with open(file,'r') as f:
    data = f.readlines()
  
```

```python
  
n = len(data)
en = np.zeros(n)
dip = np.zeros(n)
err = np.zeros(n)

for i in range(n):
    err[i] = data[i].split(',')[0]
    dip[i] = data[i].split(',')[2]
    en[i] = data[i].split(',')[3]
```

```python
fig=plt.figure(figsize=(7,5))
plt.xlabel(f'Energy [cm^{-1}]')
plt.ylabel(f'Dipole Fit Error [a.u]')
#plt.ylim([-0.1,0.1])
plt.scatter(en,err)
plt.show()
#plt.savefig('X_fit.png',dpi=500)
#plt.close()
```
